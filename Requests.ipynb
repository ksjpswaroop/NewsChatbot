```python
import requests
```

```python
r = requests.get('https://hacker-news.firebaseio.com/v0/topstories.json')
top_500 = r.json()
```

```python
print(len(top_500))

print(top_500[0])
print(top_500[-1])
```

```python
lists = []
for item in top_500:
    #get item
    r = requests.get('https://hacker-news.firebaseio.com/v0/item/' + str(item) + '.json')
    lists.append(r.json())
```

```python
print (lists[0])
```

```python
for li in lists:
    print(li['title'])
```
